```python
import pandas as pd
import matplotlib.pyplot as plt

# reading of csv file into data frame
df = pd.read_csv("FireSA.csv")

# output the data frame which is about the records of fire cases
df 
```

```python
df.isnull()
```

```python
# given that every year has the data for the months specified thus satisfying the condition, the extra null data would only bring confusion
#since those columns will no longer be necessary to satisfy the condition for the dataframe, we can drop them
df = df.dropna()

df
```

```python
#let us rename the columns to match the data they represent
df.rename(columns = {'0' : 'Year', '1' : 'Month', '2' : 'Temperature', '3' : 'Place', '4' : 'Number'}, inplace = True)
df
#be careful with handling column attributes
```

```python
#We still have the unnamed index column however this will not play a part in our data analysis so we can drop it
df.drop('Unnamed: 0', inplace = True, axis = 1)
df
#be careful with dropping columns
```

```python
df.info()
```

```python
#let us see how our dataframe looks so far before we start any data analysis
df
```

```python
#1. Which place has the most numerous cases of fire in general?
df_sv = df[df['Place'] == 'Spring Valley']
df_pv = df[df['Place'] == 'Porterville']
df_wn = df[df['Place'] == 'Winchester']
print("Spring Valley: ", df_sv['Number'].sum())
print("Portervile: ", df_pv['Number'].sum())
print("winchester: ", df_wn['Number'].sum())

#Winchester has the most numerous cases of fire 
print("\nWinchester has the most numerous cases of fire")

#let's create a visualization to better understand how high the difference is
place = {'Place' : ['Spring Valley', 'Porterville', 'Winchester'],
      'Number' : [df_sv['Number'].sum(), df_pv['Number'].sum(), df_wn['Number'].sum()]}

df_place = pd.DataFrame(place)

plt.bar(df_place['Place'], df_place['Number'])
plt.xlabel("Place")
plt.ylabel("No. of Cases")
plt.show()
```

```python
#2. What is the average temperature in Winchester when number of cases is higher than 2?

#using the previous operations and variables made in question no.1
#let's get the average temperature in the winchester dataframe with the given condition
df_wnavg = df_wn[df_wn['Number'] > 2]
avg_temp = df_wnavg['Temperature'].mean()

print("The average temperature in Winchester when number of cases is higher than 2 is: ", round(avg_temp, 2))
```

```python
#3. What is the maximum temperature during January?
df_jan = df[df['Month'] == 'January']

#let's visualize to see which year was it
plt.bar(df_jan['Year'], df_jan['Temperature'])
plt.xlabel("Year")
plt.ylabel("Temperature")
plt.show()

#since we are going to refer back to the original dataframe, take note of the removed null values to correctly emasure the index
print(" The maximum temperature during January is ", df_jan['Temperature'].max(), "recorded on the year ",
     df['Year'].iloc[df_jan['Temperature'].idxmax() - 1], "in ", df['Place'].iloc[df_jan['Temperature'].idxmax() - 1])
```

```python
#4. Which year/s has/have the most number of cases?

#similar with the dataframe creation in no.1, this one will focus on the year with its visualization
df_2002 = df[df['Year'] == 2002]
df_2003 = df[df['Year'] == 2003]
df_2004 = df[df['Year'] == 2004]
df_2005 = df[df['Year'] == 2005]
df_2006 = df[df['Year'] == 2006]
df_2007 = df[df['Year'] == 2007]
df_2008 = df[df['Year'] == 2008]
df_2009 = df[df['Year'] == 2009]

year = {'Year' : [2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009],
       'Number' : [df_2002['Number'].sum(), df_2003['Number'].sum(), df_2004['Number'].sum(), df_2005['Number'].sum(),
                  df_2006['Number'].sum(), df_2007['Number'].sum(), df_2008['Number'].sum(), df_2009['Number'].sum()]}

df_year = pd.DataFrame(year)
x = 0
plt.bar(df_year['Year'], df_year['Number'])
plt.xlabel("Year")
plt.ylabel("No. of Cases")
plt.yticks([0, 5, 10, 15, 20])
plt.show()

print("The years of 2002, 2008, and 2009 had the most cases recorded being at 15 cases per year")
```

```python
#5 Which month has the least number of cases?

#picking where we left of from the dataframe we created for january, let's also make them for the other months
#and make a visualization for it
df_feb = df[df['Month'] == 'February']
df_mar = df[df['Month'] == 'March']
df_apr = df[df['Month'] == 'April']

month = {'Month' : ['January', 'February', 'March', 'April'],
        'Number' : [df_jan['Number'].sum(), df_feb['Number'].sum(), df_mar['Number'].sum(), df_apr['Number'].sum()]}

df_month = pd.DataFrame(month)

plt.bar(df_month['Month'], df_month['Number'])
plt.xlabel("Month")
plt.ylabel("No. of Cases")
plt.show()

print("The month of April saw the most number of fire cases over the years 2002 -2009")
```

```python
#6. Which place has the greatest number of cases during April?

#although we could derive df_apr for an easier answer, using the dataframes made for the places in no.1 to derive from to make
#another dataframe would be easier in terms of visualization purposes
sv_apr = df_sv[df_sv['Month'] == 'April']
pv_apr = df_pv[df_pv['Month'] == 'April']
wn_apr = df_wn[df_wn['Month'] == 'April']

aprplace = {'Place' : ['Spring Valley', 'Porterville', 'Winchester'],
      'Number' : [sv_apr['Number'].sum(), pv_apr['Number'].sum(), wn_apr['Number'].sum()]} 

df_aprplace =  pd.DataFrame(aprplace)

plt.bar(df_aprplace['Place'], df_aprplace['Number'])
plt.xlabel("Place")
plt.ylabel("No. of Cases")
plt.show()

print("Winchester had the greatest number of cases in April with ", df_aprplace['Number'].max(), " recorded cases")
```

```python
#7. What is the least temperature during February of 2002 to 2004?

#we will derive from the february dataframe and use the min and idxmin function to get the lowest temperature 
df_feb2 = df_feb[df_feb['Year'] < 2005]
print("The lowest temperature recorded from 2002-2004 in February is ", df_feb2['Temperature'].min())
df.iloc[df_feb2['Temperature'].idxmin()]
```

```python
#8. Which place has the least temperature from 2008 to 2009?

#we will derive fromt the main dataframe, t symbolizes temperature, and again use the min and idxmin functions
dft = df[df['Year'] > 2007]

print("The place with the least temperature from 2008-2009 is ", df['Place'].iloc[dft['Temperature'].idxmin() - 3], 
     " with a temperature of ", df['Temperature'].iloc[dft['Temperature'].idxmin() - 3] )

df.iloc[dft['Temperature'].idxmin() - 3] 
# remember that since we had null values here removed, we need to count the numer of those removed rows and use that to adjust the index
```

```python
#9. What is the highest temperature from 2003 to 2007?

#similar wit no.8 except this does not ask for a place and has a different year interval

dft2 = df[df['Year'] > 2002]
dft2 = dft2[dft2['Year'] < 2008]

print("The highest temperature recorded from 2003-2007 is ", df['Temperature'].iloc[dft2['Temperature'].idxmax() -2] )

df.iloc[dft2['Temperature'].idxmax()-2] 
# remember that since we had null values here removed, we need to count the numer of those removed rows and use that to adjust the index
```
