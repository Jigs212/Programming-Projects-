```python
import pandas as pd
import matplotlib.pyplot as plt

# reading of csv file into data frame
df = pd.read_csv("grades.csv")

# output the data frame which is about the students' subject grades
df 
```

```python
#Let us find out how many null values there are
df.isnull().sum()

#as we can see math and physics had the lowest number of null values both at 1 while english and literature both have 2
```

```python
#Handling null data by filling in with average values of corresponding subjects

#filling in null values with 0 because mostly incomplete grades can insinuate no grade at all for that subject
#it cannot simply be put with average grades due to its ambiguity
df = df.fillna(0)

df
```

```python
df.info()
#now there are no null data
```

```python
df.describe()
```

```python
#1 Which subject do students excel the most? Support your answer with analysis and explanation. part 1
df.mean()

#math has the highest average among the grades comapred to the others 
```

```python
#1 Which subject do students excel the most? Support your answer with analysis and explanation. part 2
df.std() 

#math placed in at a more reliable together with physics compared to the other subjects
```

```python
#1 Which subject do students excel the most? Support your answer with analysis and explanation. part 3

# let us see which subject had the most grades equal or above 85 which is a sign of accepting honor grades in that subject
print("There are",len(df[df['Physics'] >= 85]) ,"student grades above 85 in Physics")
print("There are",len(df[df['Math'] >= 85]) ,"student grades above 85 in Math")
print("There are",len(df[df['English'] >= 85]),"student grades above 85 in English")
print("There are",len(df[df['Literature'] >= 85]) ,"student grades above 85 in Literature")

#From the results, math had the highest number of grades not below 85

#we can see from all previous tests that math has been the most consistent in terms of staying at the highest point which 
#implies that the students perform the best at math compared to other subjects 
```

```python
#2 How many students are below 87 in both Math and Physics?

#let us use a lambda expression to this question

# x here basically refers as a sort of parameter indicator for the dataframe object in order to use a comparison operation together with a logical operation
# to check for the conditions specified in the question
# the apply() function will basically apply the lambda expression to every row of the dataframe df and assign it to df2
df2 = df.apply( lambda x : True
              if x['Math'] < 87 and x['Physics'] < 87 else False, axis = 1)
#this lambda expression will basically return true for every row (which represents the students) that has grades in both math and physics
#that are below 87

# to get the number of students, the length of df2 must be obtained as it shows the students whose grades in Math and Physics are below 87
num_students = len(df2[df2 == True].index)
print("There are ", num_students, " students who have grades below 87 in both Math and Physics")

#dataframe for displaying the math and physics grade values only
df_mp = df[['Math','Physics']]

#applying the above lambda expression to this dataframe to display the student data
df_mp[df2]
```

```python
#3 What is/are the grade/s of the student/s in math who got the highest grade/s in Physics?

df_math_physics = df[['Math', 'Physics']]
#we must get the index of the student with the highest grade in physics and use it to get their grade for English using iloc
df_math_physics.iloc[[df['Physics'].idxmax()]]

# the student with the highest grade in physics(93) had a grade of 91 in math showing how proficient they are in excelling in both subject
```

```python
#4 What is/are the grade/s of the student/s in English who got the lowest grade/s in Math?

df_eng_math = df[['English','Math']]
#we must get the index of the student with the highest grade in physics and use it to get their grade for English using iloc
df_eng_math.iloc[[df['Math'].idxmin()]]

#the student with the lowest grade in math had a score of 0 due to their previous incomplete grade, their grade i english however is
# 87 which is a good grade above the subject average
```

```python
#5  How can you describe the relationship between Math and English?

#before we can start processing this we must reprocess the df dataframe to remove null values instead of filling them in as 0
# so that we can get more consistent results

# reading of csv file into data frame
# csv file is set in documents folder of the original creator of this ipnyb file (Sanjay Sajnani)
df = pd.read_csv("C://Users//Jigs//Documents//python prac//grades.csv")

df = df.dropna()

#this is for setting the size of the figure
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(8)

plt.scatter(df[['Math']],df[['English']])
plt.xlabel("Math Grades") #x-axis label
plt.ylabel("Engish Grades") #y-axis label
plt.title("Relationship Between Students' Grades in Math and English")

# we can see that the subjects English and Math have a negative correlation due to the plot results
```
